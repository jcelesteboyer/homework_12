# Lambda Functions

A lambda function is an anonymous function, which means it doesnt have a name.

A lambda function has less code than a normal function. 

We use lambdas as functions, to filter results, apply results to a column or row, call more complicated functions, etc.

```python
#https://www.analyticsvidhya.com/blog/2020/03/what-are-lambda-functions-in-python/
import pandas as pd

jeopardy_df = pd.read_csv("../week_05/JEOPARDY_CSV.csv")
jeopardy_df.head()
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.replace(" ","")
jeopardy_df.head()
```

```python
jeopardy_df.dtypes
```

```python
jeopardy_df['AirDate']=pd.to_datetime(jeopardy_df['AirDate'],format='%m/%d/%Y')
jeopardy_df.dtypes
```

```python
#assign is typically used to assign a new columns
jeopardy_df = jeopardy_df.assign(month=lambda x: x['AirDate'].dt.month)
jeopardy_df.head()
```

```python
# how do we do the same thing that we did in the cell above but using apply?
jeopardy_df['month2'] = jeopardy_df.apply(lambda x: x['AirDate'].month, axis=1)
jeopardy_df.head()
```

# Filter

```python
value_list = jeopardy_df['Value'].tolist()
value_list=value_list[0:50]
value_list

#goal = want a list of all values greater than 300
```

```python
#we can either include all logic in the lambda, OR we can replace
#the characters we dont want and then run the lambda to filter
#clean_up_lambda=lambda num:int(num.replace('$','').replace(' ','').replace(',',''))

filtered_list = list(filter(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))

filtered_list
```

# Map and Reduce

Map allows us to apply a function over all the input values. It allows us to iterate over various sizes of inputs: whole dataframe, column, list.

Reduce performs a computation or aggregation over the whole list. 


```python
mapped_list = list(map(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300,value_list))
mapped_list
```
